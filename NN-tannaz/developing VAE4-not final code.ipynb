```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Excel file
file_path = r'C:\Users\TanMeh\Desktop\jupyter\to code - Copy.xlsx'  # Replace with the actual path to your file
excel_data = pd.ExcelFile(file_path)
df = excel_data.parse('Sheet1')
```

```python
# Cell 3: Define Column Pairs to Compare
columns_to_compare = [
    ('P_SEXE', 'P_SEXE-2'),
    ('P_GRAGE', 'P_GRAGE-2'),
    ('D_ORIXCOOR', 'D_ORIXCOOR-2'),
    ('D_ORIYCOOR', 'D_ORIYCOOR-2'),
    ('D_DESTXCOOR', 'D_DESTXCOOR-2'),
    ('D_DESTYCOOR', 'D_DESTYCOOR-2'),
    ('D_HREDE', 'D_HREDE-2')
]
```

```python
# Cell 4: Define Function for Frequency Distribution Comparison
def plot_frequency_distributions(df, column_pairs):
    """
    Generate histograms to compare frequency distributions of specified column pairs.
    
    Parameters:
    - df: DataFrame containing the data.
    - column_pairs: List of tuples containing pairs of column names to compare.
    """
    for col1, col2 in column_pairs:
        plt.figure(figsize=(12, 6))
        
        # Plot histogram for the first column
        plt.hist(df[col1], bins=30, alpha=0.5, label=f'{col1}')
        # Plot histogram for the second column
        plt.hist(df[col2], bins=30, alpha=0.5, label=f'{col2}')
        
        # Add titles and labels
        plt.title(f'Frequency Distribution Comparison: {col1} vs {col2}')
        plt.xlabel('Value')
        plt.ylabel('Frequency')
        plt.legend()
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
```

```python
# Cell 5: Plot Frequency Distributions
plot_frequency_distributions(df, columns_to_compare)
```
